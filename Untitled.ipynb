```python
def Isnull(dataframe):
    se1 = dataframe.isnull().sum()
    se1 = se1[se1>0]
    se1.sort_values(inplace = True)
    print('缺失值还有这么多个:',len(se1))
    return se1
```

```python
def IsnullList(dataframe):
    se1 = dataframe.isnull().sum()
    se1 = se1[se1>0]
    se1.sort_values(inplace = True)
    print('缺失值的columns如下:')
    return list(se1.index)
```

```python
train['Electrical'].isnull()[train['Electrical'].isnull()==True]
```

```python
def outliersIndex(dataframe,column):
    return dataframe[column].isnull()[dataframe[column].isnull()==True]
```

```python
l1 = IsnullList(train)
```

```python
numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
for i in l1:
    if train[i].dtype in numeric_dtypes:
        print('quantitative:',i)
    else:
        print('object:',i)
```
